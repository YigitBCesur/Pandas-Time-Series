```python
#@title 31-day period from the desired date on a DAILY frequency ( istenilen tarihten 31 gunluk periyod )
pd.date_range(start='2020-01-01',periods =31 , freq='D')

# important = when we use "periods", we do not use 'end' date - 'periods' eklendiginde "end" yok.
```

```python
#@title 10-week period from the desired date on a WEEKLY frequency ( istenilen tarihten 10 haftalik periyod )
pd.date_range(start='2020-01-01',periods =10 , freq='W')

# important = when we use "periods", we do not use 'end' date - 'periods' eklendiginde "end" yok.
```

```python
#@title MONTHLY period from the desired date on a B-MONTHLY  frequency  ( aylik olarak istenilen tarihlerde   B=Month-Her ayin basina gore )
pd.date_range(start='2020-01-01', end='2022-12-01', freq='B')
```

```python
pd.date_range(start='2020-01-01', end='2022-12-01', freq='2B')
```

```python
#@title 10 periods monthly UNTIL (end) the desired date (M=Month) - 10 period aylik olarak istenilen tarihe kadar   ( M=Month )
pd.date_range(end='2020-12-31', periods=10, freq='M')
```

```python
#@title 10 periods weekly until the desired date (W=Week, H=Hour, M=Month, Y=Year, MS=Month Start)
#10 period haftalik olarak istenilen tarihe kadar   ( W=Week , H=Hour , B=Month , Y=Year , MS=Month Start )

pd.date_range(end='2020-12-31', periods=10, freq='MS')    # otomatik olarak Mon-Sun ayarlar
```

```python
#@title From the desired date Only 10 periods just (Weekly) Wednesdays
#istenilen tarihten Sadece 10 period Haftalik Carsamba gunleri
pd.date_range(start='2020-12-31', periods=10, freq='w-Wed')
```

```python
#@title Starting from the desired date for 5 periods, from the the start of THE MONTH (MS=Month Start)
#5 period olarak istenilen tarihten itibaren AY BASLARIndan    ( MS=Month Start )
pd.date_range(start='2020-01-01', periods=5, freq='MS')
```

```python
#@title 10 periods, 2 months apart from the desired date
#10 period olarak istenilen tarihten itibaren iki ay arayla
pd.date_range(start='2020-01-01', periods=10, freq=pd.DateOffset(months=2))
```

* Down Sampling â€“ Going from a fine time grain to a lower one. Example: Months to Years.

Pandas Resample transforms your time series data into different frequencies. Think of it like a group function, but for time series data.

```python
#@title Convert to YEAR END with MEAN --- YIL SONLARINA olarak donustur ve ORTALAMASINI al
df.resample('Y').mean()
```

```python
#@title Convert to The Start of Year  with MEAN --- YIL BASLARINA baslayacak sekilde donustur ve ORTALAMASINI al
df.resample('YS').mean()
```

```python
#@title Convert to WEEKLY PERIODS with MEAN --- HAFTALIK PERIYODLAR olarak donustur ve ORTALAMASINI al   ( Q,Y,A,M )
df_w=df.resample('W', kind='period').mean()
df_w
```

# **`Advanced Indexing with reindex()`**

```python
df_d=df.resample('D').mean()
df_d
```

```python
birthd= pd.date_range(end='2018-12-24', periods=10, freq=pd.DateOffset(years=1))
birthd
```

```python
df_d.loc[birthd]  #ERROR
```

```python
df_d.reindex(birthd , method='bfill')
```

```python
#@title with all 31 days in May 2015
alldays_may =pd.date_range( start="2015-05-01" , periods=31 , freq="D")
alldays_may
```

```python
#@title alldays_may   FORWARD FILL any NaN values!
df.reindex(alldays_may, method='ffill')
```
